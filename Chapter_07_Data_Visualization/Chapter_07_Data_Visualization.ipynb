# Static 2D Plotting

**One Dimensional Data Set**

```python
# Import the numpy library for numerical operations.
import numpy as np
```

```python
# Set the seed for the random number generator to ensure reproducibility.
np.random.seed(1000)
```

```python
# Generate an array of 20 random numbers from a standard normal distribution.
y = np.random.standard_normal(20)
```

```python
# Create an array x with indices corresponding to the length of y.
x = np.arange(len(y))
# Create a basic line plot of y against x.
plt.plot(x, y);
```

```python
# Plot the cumulative sum of the array y.
plt.plot(y.cumsum());
```

```python
# Plot the cumulative sum of the array y.
plt.plot(y.cumsum())
# Turn off the grid lines in the plot.
plt.grid(False);
```

```python
# Plot the cumulative sum of the array y.
plt.plot(y.cumsum())
# Set the x-axis limits from -1 to 20.
plt.xlim(-1, 20)
# Set the y-axis limits based on the minimum and maximum values of the cumulative sum, with some padding.
plt.ylim(np.min(y.cumsum()) -1,
         np.max(y.cumsum()) + 1);
```

```python
# Create a new figure with a specified size (width=10 inches, height=6 inches).
plt.figure(figsize=(10, 6))
# Plot the cumulative sum of y as a blue line with a width of 1.5.
plt.plot(y.cumsum(), 'b', lw=1.5)
# Plot the cumulative sum of y again, this time as red circles at each data point.
plt.plot(y.cumsum(), 'ro')
# Turn on the grid lines.
plt.grid(True)
# Set the plot limits to be 'tight' around the data.
plt.axis('tight')
# Label the x-axis as 'index'.
plt.xlabel('index')
# Label the y-axis as 'value'.
plt.ylabel('value')
# Set the title of the plot to 'A Simple Plot'.
plt.title('A Simple Plot');
```

#Two - Dimensional Data Set

```python
# Generate a 2D array of 20x2 random numbers from a standard normal distribution and calculate the cumulative sum along the columns.
y = np.random.standard_normal((20, 2)).cumsum(axis=0)
```

```python
# Create a new figure with a specified size (width=10 inches, height=6 inches).
plt.figure(figsize=(10, 6))
# Plot the 2D array y as lines with a width of 1.5.
plt.plot(y, lw=1.5)
# Plot the 2D array y again, this time as red circles at each data point.
plt.plot(y, 'ro')
# Label the x-axis as 'index'.
plt.xlabel('index')
# Label the y-axis as 'value'.
plt.ylabel('value')
# Set the title of the plot to 'A Simple Plot'.
plt.title('A Simple Plot');
```

```python
# Create a new figure with a specified size (width=10 inches, height=6 inches).
plt.figure(figsize=(10, 6))
# Plot the first column of y as a line with a width of 1.5 and label it '1st'.
plt.plot(y[:, 0], lw=1.5, label='1st')
# Plot the second column of y as a line with a width of 1.5 and label it '2nd'.
plt.plot(y[:, 1], lw=1.5, label='2nd')
# Plot all data points in y as red circles.
plt.plot(y, 'ro')
# Add a legend to the plot and place it at the best location (loc=0).
plt.legend(loc = 0)
# Label the x-axis as 'index'.
plt.xlabel('index')
# Label the y-axis as 'value'.
plt.ylabel('value')
# Set the title of the plot to 'A Simple Plot'.
plt.title('A Simple Plot');
```

```python
# Create a figure and a set of subplots, where ax1 is the primary axes.
fig, ax1 = plt.subplots()
# Plot the first column of y on ax1 as a blue line with a width of 1.5 and label it '1st'.
plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
# Plot the data points of the first column of y on ax1 as red circles.
plt.plot(y[:, 0], 'ro')
# Add a legend to ax1 and place it at location 8 (lower center).
plt.legend(loc = 8)
# Label the x-axis of ax1 as 'index'.
plt.xlabel('index')
# Label the y-axis of ax1 as 'value 1st'.
plt.ylabel('value 1st')
# Set the title of the plot to 'A Simple Plot'.
plt.title('A Simple Plot');

# Create a second axes that shares the same x-axis as ax1.
ax2 = ax1.twinx()
# Plot the second column of y on ax2 as a green line with a width of 1.5 and label it '2nd'.
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
# Plot the data points of the second column of y on ax2 as red circles.
plt.plot(y[:, 1], 'ro')
# Add a legend to ax2 and place it at the best location (loc=0).
plt.legend(loc = 0)
# Label the y-axis of ax2 as 'value 2nd'.
plt.ylabel('value 2nd');
```

# Other Plot Styles

```python
# Import necessary libraries: numpy for numerical operations and matplotlib.pyplot for plotting.
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate a 2D array of 1000x2 random numbers from a standard normal distribution.
y = np.random.standard_normal((1000, 2))
```

```python
# Create a new figure with a specified size and generate a scatter plot of the first and second columns of y.
plt.figure(figsize=(10, 6))
plt.plot(y[:, 0], y[:, 1], 'ro') # 'ro' specifies red circles as markers
plt.xlabel('1st') # Label the x-axis
plt.ylabel('2nd') # Label the y-axis
plt.title('Scatter Plot'); # Set the title of the plot
```

```python
# Create a new figure with a specified size and generate a scatter plot with default blue circle markers.
plt.figure(figsize=(10, 6))
plt.plot(y[:, 0], y[:, 1], 'o') # 'o' specifies default blue circles as markers
plt.xlabel('1st') # Label the x-axis
plt.ylabel('2nd'); # Label the y-axis
plt.title('Scatter Plot'); # Set the title of the plot
```

```python
# Generate an array of 1000 random integers between 0 and 9 (inclusive) to be used for coloring the scatter plot points.
c = np.random.randint(0, 10, len(y))
```

```python
# Create a new figure and generate a scatter plot with points colored according to the 'c' array and sized at 50.

plt.figure(figsize=(10, 6))

plt.scatter(y[:, 0], y[:, 1], c=c, s=50, cmap='coolwarm',marker='o') # 'c' for color, 's' for size, 'cmap' for colormap, 'marker' for marker style

plt.colorbar() # Add a colorbar to indicate the mapping of colors to values in 'c'
plt.xlabel('1st') # Label the x-axis

plt.ylabel('2nd') # Label the y-axis

plt.title('Scatter Plot'); # Set the title of the plot
```

```python
# Create a new figure and generate a histogram of the two columns of y.
plt.figure(figsize=(10, 6))

plt.hist(y, label=['1st', '2nd'], bins=25) # 'label' for legend entries, 'bins' for the number of histogram bins

plt.legend(loc=0) # Add a legend at the best location

plt.xlabel('value') # Label the x-axis

plt.ylabel('frequency') # Label the y-axis

plt.title('Histogram'); # Set the title of the plot
```

```python
# Create a new figure and generate a stacked histogram with specified colors, bins, and transparency.

plt.figure(figsize=(10, 6))

plt.hist(y, label=['1st', '2nd'], color=['b', 'g'], # 'color' for specifying colors for each data set

         stacked=True, bins=20, alpha=0.5) # 'stacked' for stacking the histograms, 'bins' for the number of bins, 'alpha' for transparency

plt.legend(loc=0) # Add a legend at the best location

plt.xlabel('value') # Label the x-axis

plt.ylabel('frequency') # Label the y-axis

plt.title('Histogram'); # Set the title of the plot
```

```python
# Create a figure and axes for a boxplot, then generate the boxplot of the two columns of y.

fig, ax = plt.subplots(figsize=(10, 6)) # Create a figure and an axes object

ax.boxplot(y) # Generate the boxplot on the axes

plt.setp(ax, xticklabels=['1st', '2nd']) # Set the labels for the x-axis ticks

plt.xlabel('data set') # Label the x-axis

plt.ylabel('value') # Label the y-axis

plt.title('Boxplot'); # Set the title of the plot
```

```python
# Define a function and generate data points for plotting a curve and shading an area under the curve.
def func(x):

    # Define a function for which we will calculate and plot values.
    return 0.5 * np.exp(x) + 1

a, b = 0.5, 1.5 # Define the limits for the shaded area.

x = np.linspace(0, 2) # Generate evenly spaced values for the x-axis for the curve.

y = func(x) # Calculate the corresponding y-values using the defined function.

Ix = np.linspace(a, b) # Generate evenly spaced values for the x-axis within the integration limits.

Iy = func(Ix) # Calculate the corresponding y-values for the integration limits.

verts = [(a, 0)] + list(zip(Ix, Iy)) + [(b, 0)] # Create a list of vertices to define the polygon for the shaded area.
```

```python
# Import the Polygon patch and create a plot with a curve, a shaded area representing an integral, and labels with mathematical notation.

from matplotlib.patches import Polygon # Import the Polygon class for creating polygons.

fig, ax = plt.subplots(figsize=(10, 6)) # Create a figure and an axes object for plotting.

plt.plot(x, y, 'b', linewidth = 2) # Plot the curve defined by x and y as a blue line with a width of 2.

plt.ylim(ymin=0) # Set the lower limit of the y-axis to 0.

poly = Polygon(verts, facecolor='0.7', edgecolor='0.5') # Create a polygon patch using the defined vertices, with a gray face color and edge color.

ax.add_patch(poly) # Add the polygon patch to the axes.

plt.text(0.5 * (a + b), 1, r'$\int_a^b f(x)\mathrm{d}x$', # Add text to the plot representing the integral.

         horizontalalignment='center', fontsize=20) # Set horizontal alignment and font size for the text.

plt.figtext(0.9, 0.075, '$x$') # Add text for the x-axis label using figure coordinates.

plt.figtext(0.075, 0.9, '$f(x)$') # Add text for the y-axis label using figure coordinates.

ax.set_xticks((a, b)) # Set the x-axis ticks at positions a and b.

ax.set_xticklabels(('$a$', '$b$')) # Set the labels for the x-axis ticks.

ax.set_yticks([func(a), func(b)]) # Set the y-axis ticks at the function values at a and b.

ax.set_yticklabels(('$f(a)$', '$f(b)$')) # Set the labels for the y-axis ticks.
```
