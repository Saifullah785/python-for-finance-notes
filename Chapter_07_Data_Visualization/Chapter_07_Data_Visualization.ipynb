# Static 2D Plotting

**One Dimensional Data Set**

```python
# Import the numpy library for numerical operations.
import numpy as np
```

```python
# Set the seed for the random number generator to ensure reproducibility.
np.random.seed(1000)
```

```python
# Generate an array of 20 random numbers from a standard normal distribution.
y = np.random.standard_normal(20)
```

```python
# Create an array x with indices corresponding to the length of y.
x = np.arange(len(y))
# Create a basic line plot of y against x.
plt.plot(x, y);
```

```python
# Plot the cumulative sum of the array y.
plt.plot(y.cumsum());
```

```python
# Plot the cumulative sum of the array y.
plt.plot(y.cumsum())
# Turn off the grid lines in the plot.
plt.grid(False);
```

```python
# Plot the cumulative sum of the array y.
plt.plot(y.cumsum())
# Set the x-axis limits from -1 to 20.
plt.xlim(-1, 20)
# Set the y-axis limits based on the minimum and maximum values of the cumulative sum, with some padding.
plt.ylim(np.min(y.cumsum()) -1,
         np.max(y.cumsum()) + 1);
```

```python
# Create a new figure with a specified size (width=10 inches, height=6 inches).
plt.figure(figsize=(10, 6))
# Plot the cumulative sum of y as a blue line with a width of 1.5.
plt.plot(y.cumsum(), 'b', lw=1.5)
# Plot the cumulative sum of y again, this time as red circles at each data point.
plt.plot(y.cumsum(), 'ro')
# Turn on the grid lines.
plt.grid(True)
# Set the plot limits to be 'tight' around the data.
plt.axis('tight')
# Label the x-axis as 'index'.
plt.xlabel('index')
# Label the y-axis as 'value'.
plt.ylabel('value')
# Set the title of the plot to 'A Simple Plot'.
plt.title('A Simple Plot');
```

#Two - Dimensional Data Set

```python
# Generate a 2D array of 20x2 random numbers from a standard normal distribution and calculate the cumulative sum along the columns.
y = np.random.standard_normal((20, 2)).cumsum(axis=0)
```

```python
# Create a new figure with a specified size (width=10 inches, height=6 inches).
plt.figure(figsize=(10, 6))
# Plot the 2D array y as lines with a width of 1.5.
plt.plot(y, lw=1.5)
# Plot the 2D array y again, this time as red circles at each data point.
plt.plot(y, 'ro')
# Label the x-axis as 'index'.
plt.xlabel('index')
# Label the y-axis as 'value'.
plt.ylabel('value')
# Set the title of the plot to 'A Simple Plot'.
plt.title('A Simple Plot');
```

```python
# Create a new figure with a specified size (width=10 inches, height=6 inches).
plt.figure(figsize=(10, 6))
# Plot the first column of y as a line with a width of 1.5 and label it '1st'.
plt.plot(y[:, 0], lw=1.5, label='1st')
# Plot the second column of y as a line with a width of 1.5 and label it '2nd'.
plt.plot(y[:, 1], lw=1.5, label='2nd')
# Plot all data points in y as red circles.
plt.plot(y, 'ro')
# Add a legend to the plot and place it at the best location (loc=0).
plt.legend(loc = 0)
# Label the x-axis as 'index'.
plt.xlabel('index')
# Label the y-axis as 'value'.
plt.ylabel('value')
# Set the title of the plot to 'A Simple Plot'.
plt.title('A Simple Plot');
```

```python
# Multiply the values in the first column of y by 100.
y[:, 0] = y[:, 0] * 100
```

```python
# Create a new figure with a specified size (width=10 inches, height=6 inches).
plt.figure(figsize=(10, 6))
# Plot the first column of y as a line with a width of 1.5 and label it '1st'.
plt.plot(y[:, 0], lw=1.5, label='1st')
# Plot the second column of y as a line with a width of 1.5 and label it '2nd'.
plt.plot(y[:, 1], lw=1.5, label='2nd')
# Plot all data points in y as red circles.
plt.plot(y, 'ro')
# Add a legend to the plot and place it at the best location (loc=0).
plt.legend(loc = 0)
# Label the x-axis as 'index'.
plt.xlabel('index')
# Label the y-axis as 'value'.
plt.ylabel('value')
# Set the title of the plot to 'A Simple Plot'.
plt.title('A Simple Plot');
```

```python
# Create a figure and a set of subplots, where ax1 is the primary axes.
fig, ax1 = plt.subplots()
# Plot the first column of y on ax1 as a blue line with a width of 1.5 and label it '1st'.
plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
# Plot the data points of the first column of y on ax1 as red circles.
plt.plot(y[:, 0], 'ro')
# Add a legend to ax1 and place it at location 8 (lower center).
plt.legend(loc = 8)
# Label the x-axis of ax1 as 'index'.
plt.xlabel('index')
# Label the y-axis of ax1 as 'value 1st'.
plt.ylabel('value 1st')
# Set the title of the plot to 'A Simple Plot'.
plt.title('A Simple Plot');

# Create a second axes that shares the same x-axis as ax1.
ax2 = ax1.twinx()
# Plot the second column of y on ax2 as a green line with a width of 1.5 and label it '2nd'.
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
# Plot the data points of the second column of y on ax2 as red circles.
plt.plot(y[:, 1], 'ro')
# Add a legend to ax2 and place it at the best location (loc=0).
plt.legend(loc = 0)
# Label the y-axis of ax2 as 'value 2nd'.
plt.ylabel('value 2nd');
```

```python
# Create a new figure with a specified size (width=10 inches, height=6 inches).
plt.figure(figsize=(10, 6))
# Create the first subplot in a 2x1 grid (2 rows, 1 column, first plot).
plt.subplot(211)
# Plot the first column of y on the first subplot as a line with a width of 1.5 and label it '1st'.
plt.plot(y[:, 0], lw=1.5, label='1st')
# Plot the data points of the first column of y on the first subplot as red circles.
plt.plot(y[:, 0], 'ro')
# Add a legend to the first subplot and place it at the best location (loc=0).
plt.legend(loc=0)
# Label the y-axis of the first subplot as 'value'.
plt.ylabel('value')
# Set the title of the first subplot to 'A Simple Plot'.
plt.title('A Simple Plot')
# Create the second subplot in a 2x1 grid (2 rows, 1 column, second plot).
plt.subplot(212)
# Plot the second column of y on the second subplot as a green line with a width of 1.5 and label it '2nd'.
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
# Plot the data points of the second column of y on the second subplot as red circles.
plt.plot(y[:, 1], 'ro')
# Add a legend to the second subplot and place it at the best location (loc=0).
plt.legend(loc=0)
# Label the x-axis of the second subplot as 'index'.
plt.xlabel('index')
# Label the y-axis of the second subplot as 'value'.
plt.ylabel('value');
```

```python
# Create a new figure with a specified size (width=10 inches, height=6 inches).
plt.figure(figsize=(10, 6))
# Create the first subplot in a 1x2 grid (1 row, 2 columns, first plot).
plt.subplot(121)
# Plot the first column of y on the first subplot as a line with a width of 1.5 and label it '1st'.
plt.plot(y[:, 0], lw=1.5, label='1st')
# Plot the data points of the first column of y on the first subplot as red circles.
plt.plot(y[:, 0], 'ro')
# Add a legend to the first subplot and place it at the best location (loc=0).
plt.legend(loc=0)
# Label the x-axis of the first subplot as 'index'.
plt.xlabel('index')
# Label the y-axis of the first subplot as 'value'.
plt.ylabel('value')
# Set the title of the first subplot to '1st Data Set'.
plt.title('1st Data Set')
# Create the second subplot in a 1x2 grid (1 row, 2 columns, second plot).
plt.subplot(122)
# Create a bar plot of the second column of y on the second subplot.
# The x-axis values are indices from 0 to the length of y minus 1.
# The bars have a width of 0.5, are green, and are labeled '2nd'.
plt.bar(np.arange(len(y)), y[:, 1], width=0.5,
        color='g', label='2nd')
# Add a legend to the second subplot and place it at the best location (loc=0).
plt.legend(loc=0)
# Label the x-axis of the second subplot as 'index'.
plt.xlabel('index')
# Set the title of the second subplot to '2nd Data Set'.
plt.title('2nd Data Set');
```
