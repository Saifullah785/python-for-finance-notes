# **Data Analysis with pandas**

pandas Basics

**First Steps with DataFrame Class**

```python
# Import the pandas library, commonly aliased as 'pd'
import pandas as pd
```

```python
# Create a pandas DataFrame from a list of numbers
df  = pd.DataFrame([10, 20, 30, 40],
                  columns=['numbers'],
                  index=['a', 'b', 'c', 'd'])
```

```python
# Display the DataFrame
df
```

```python
# Access and display the index of the DataFrame
df.index
```

```python
# Access and display the columns of the DataFrame
df.columns
```

```python
# Access a row by its label using .loc[]
df.loc['c']
```

```python
# Access multiple rows by their labels using .loc[]
df.loc[['a','d']]
```

```python
# Access rows by their integer position using .iloc[] (slice)
df.iloc[1:3]
```

```python
# Calculate the sum of the columns in the DataFrame
df.sum()
```

```python
# Apply a function (squaring each element) to the DataFrame using .apply()
df.apply(lambda x: x**2)
```

```python
# Square each element in the DataFrame using element-wise operation
df ** 2
```

```python
# Add a new column 'floats' to the DataFrame
df['floats'] = (1.5, 2.5, 3.5, 4.5)
```

```python
# Display the DataFrame with the new column
df
```

```python
# Access and display the 'floats' column (as a Series)
df['floats']
```

```python
# Add another new column 'names' using a pandas Series with a specified index
df['names'] = pd.Series(['Yves', 'Sandra', 'Lilli', 'Henry'], index=['d', 'a', 'b', 'c'])
```

```python
# Display the DataFrame with the 'names' column
df
```

```python
# Concatenate the current DataFrame with a new DataFrame containing one row, ignoring the original index
pd.concat((df, pd.DataFrame({'numbers': 100, 'floats': 5.75, 'names': 'Jil'}, index=[0,])),
               ignore_index=True)
```

```python
# Assign the result of the concatenation back to the original DataFrame variable
df = pd.concat((df, pd.DataFrame({'numbers': 100, 'floats': 5.75, 'names': 'Jil'}, index=[0,])),
               ignore_index=True)
```

```python
# Display the updated DataFrame after concatenation
df
```

```python
# Add a new row with label 'z' and some values. Note that missing values will be NaN.
df.loc['z'] = {'names': 'Liz'}
```

```python
# Display the DataFrame with the newly added row
df
```

```python
# Display the data types of each column in the DataFrame
df.dtypes
```

```python
# Calculate the mean of the 'numbers' and 'floats' columns
df[['numbers', 'floats']].mean()
```

```python
# Calculate the standard deviation of the 'numbers' and 'floats' columns
df[['numbers', 'floats']].std()
```

#**Group By Operations**

```python
# Add a 'Quarter' column to the DataFrame to categorize data by quarter
df ['Quarter'] = ['Q1', 'Q1', 'Q1', 'Q2', 'Q2', 'Q2', 'Q3', 'Q3', 'Q3']

# Display the DataFrame with the new 'Quarter' column
df
```

```python
# Group the DataFrame by the 'Quarter' column
groups = df.groupby('Quarter')
```

```python
# Get the size of each group (number of rows in each quarter)
groups.size()
```

```python
# Calculate the mean of each group
groups.mean()
```

```python
# Get the maximum value within each group
groups.max()
```

```python
# Aggregate data in each group using multiple functions ('min' and 'max') and round the result to 2 decimal places
groups.aggregate(['min', 'max']).round(2)
```

```python
# Add an 'Odd_Even' column to the DataFrame for further grouping
df['Odd_Even'] = ['Odd', 'Even', 'Odd', 'Even', 'Odd', 'Even', 'Odd', 'Even', 'Odd']
```

```python
# Group the DataFrame by both 'Quarter' and 'Odd_Even'
groups = df.groupby(['Quarter', 'Odd_Even'])
```

```python
# Get the size of each sub-group
groups.size()
```

```python
# Aggregate the 'No1' and 'No2' columns of each group with 'sum' and 'mean'
groups[['No1','No2']].aggregate(['sum', 'mean'])
```

#**Complex Selection**

```python
# Generate a 10x2 NumPy array of random numbers from a standard normal distribution
data = np.random.standard_normal((10, 2))
```

```python
# Create a pandas DataFrame from the generated data with columns 'x' and 'y'
df = pd.DataFrame(data, columns=[ 'x', 'y'])
```

```python
# Display a concise summary of the DataFrame
df.info()
```

```python
# Display the first 5 rows of the DataFrame
df.head()
```

```python
# Display the last 5 rows of the DataFrame
df.tail()
```

```python
# Create a boolean Series indicating where values in the 'x' column are greater than 0.5
df['x'] > 0.5
```

```python
# Create a boolean Series indicating where 'x' is positive and 'y' is negative
(df['x']>0) & (df['y']<0)
```

```python
# Create a boolean Series indicating where 'x' is positive or 'y' is negative
(df['x']>0) | (df['y']<0)
```

```python
# Select and display rows where the value in the 'x' column is positive
df[df['x']>0]
```

```python
# Use the query method to select and display rows where 'x' is positive
df.query('x>0')
```

```python
# Select and display rows where 'x' is positive and 'y' is negative
df[(df['x']>0) & (df['y']<0)]
```

```python
# Use the query method to select and display rows where 'x' is positive and 'y' is negative
df.query('x>0 and y<0')
```

```python
# Select and display rows where 'x' is positive or 'y' is negative
df[(df['x']>0) | (df['y']<0)]
```

```python
# Create a boolean DataFrame indicating which elements are positive
df > 0
```

```python
# Select and display elements from the DataFrame that are positive, replacing non-positive values with NaN
df[df > 0]
```
